Detecting Outliers using concept of IQR

```python
def remove_outliers(df):
    outliers = {}
    for col in df.columns:
        if str(df[col].dtype) != 'object':
            df = df[np.abs(df[col]-df[col].mean()) < (3*df[col].std())]
            olrs = df[~(np.abs(df[col]-df[col].mean()) < (3*df[col].std()))]
            outliers = pd.DataFrame(olrs)
    
    return print('These outliers have been removed from your dataset: \n\n', outliers,
                'New shape is:', df.shape)
```

Hypothesis Testing


“Is there any seasonality in number of customers using the cab service?”

```python
from statsmodels.tsa.stattools import adfuller


yellow_customers_per_month = data[data['Company'] == 'Yellow Cab']['Customer ID'].resample('M').count()

# Perform the Augmented Dickey-Fuller test for Yellow Cab
yellow_result = adfuller(yellow_customers_per_month)
yellow_p_value = yellow_result[1]
print('P value is ', yellow_p_value)


# Check for stationarity for Yellow Cab
if yellow_p_value < 0.05:
    print("Yellow Cab: The time series is stationary. There might be seasonality.")
else:
    print("Yellow Cab: The time series is not stationary. Seasonality might not be present.")
```

```python
from statsmodels.tsa.stattools import adfuller
import pandas as pd

# Assuming 'data' is your DataFrame containing the relevant data

# Convert the 'Invoice Date' column to a datetime format if it's not already
data['Date of Travel'] = pd.to_datetime(data['Date of Travel'])

# Set the DataFrame index to the 'Invoice Date' column
data.set_index('Date of Travel', inplace=True)

# Extract the number of customers per month for Pink Cab
pink_customers_per_month = data[data['Company'] == 'Pink Cab']['Customer ID'].resample('M').count()

# Perform the Augmented Dickey-Fuller test for Pink Cab
pink_result = adfuller(pink_customers_per_month)
pink_p_value = pink_result[1]

# Print the p-value
print("P value for Pink Cab:", pink_p_value)

# Check for stationarity for Pink Cab
if pink_p_value < 0.05:
    print("Pink Cab: The time series is stationary. There might be seasonality.")
else:
    print("Pink Cab: The time series is not stationary. Seasonality might not be present.")
```

Hypothesis 2: Is there any difference in profit regarding Gender <br>

H0: there is no difference in profit regarding gender <br>
H1: there is a difference in profit regarding gender

```python
# Filter data for Yellow Cab and Male customers
from scipy import stats

yellow_male_profit = data[(data['Company'] == 'Yellow Cab') & (data['Gender'] == 'Male')]['Profit']

# Filter data for Yellow Cab and Female customers
yellow_female_profit = data[(data['Company'] == 'Yellow Cab') & (data['Gender'] == 'Female')]['Profit']

# Perform the independent t-test
yellow_t_statistic, yellow_p_value = stats.ttest_ind(yellow_male_profit, yellow_female_profit, equal_var=True)
print('p value is:', yellow_p_value)

# Interpret the results for Yellow Cab
if yellow_p_value < 0.05:
    print("For Yellow Cab: We accept alternative hypothesis (H1) that there is a difference in profit regarding gender.")
else:
    print("For Yellow Cab: We accept null hypothesis (H0) that there is no difference in profit regarding gender.")
```

```python
# Filter data for Pink Cab and Male customers
pink_male_profit = data[(data['Company'] == 'Pink Cab') & (data['Gender'] == 'Male')]['Profit']

# Filter data for Pink Cab and Female customers
pink_female_profit = data[(data['Company'] == 'Pink Cab') & (data['Gender'] == 'Female')]['Profit']

# Perform the independent t-test
pink_t_statistic, pink_p_value = stats.ttest_ind(pink_male_profit, pink_female_profit, equal_var=True)
print('p value is:', pink_p_value)

# Interpret the results for Pink Cab
if pink_p_value < 0.05:
    print("For Pink Cab: We accept alternative hypothesis (H1) that there is a difference in profit regarding gender.")
else:
    print("For Pink Cab: We accept null hypothesis (H0) that there is no difference in profit regarding gender.")
```
