```python
num_features = data.select_dtypes(exclude='object')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming num_features is your correlation matrix
correlation_matrix = num_features.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(20, 8))

# Create a heatmap with the mask applied
sns.heatmap(
    correlation_matrix,
    mask=mask,  # Apply the mask
    annot=True,
    cmap='coolwarm',  # You can choose the colormap you prefer
    fmt=".2f",  # Format for annotating values
)

plt.show()
```
